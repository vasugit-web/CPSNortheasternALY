```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_excel( 'C:/naraharitransactions/AddmissionsInfo/ALY6140 72241 Analytics Systems Technology SEC 11 Fall 2019 CPS/Week4/civ6_players.xlsx')
##df.head(50)
```

```python
##Q1.Select TOP LEADERMOSTSELECTED based on  Sum of ACTIVEDAYS he choosen by different player?
import matplotlib.pyplot as plt
df2 = df.groupby('LEADERMOSTSELECTED').ACTIVEDAYS.sum().sort_values('index',ascending = False).head(10)
figbar =plt.figure()
ax1 = figbar.add_subplot(1,1,1)
ax1.set_xlabel('LEADERMOSTSELECTED')
ax1.set_ylabel('Sum of ACTIVEDAYS')
ax1.set_title("TOP LEADERMOSTSELECTED wise Sum of ACTIVEDAYS")
plt.plot(df2)
plt.xticks(rotation=90)
plt.margins(0.2)
plt.subplots_adjust(bottom=0.15)
plt.show()

```

```python
##Question2 Civilization seleted variation with active days ?
import seaborn as sns 
sns.violinplot(df2['CIVILIZATIONMOSTSELECTED'])##, df2['CIVILIZATIONMOSTSELECTED']) 
sns.despine()
```

```python
##Question 3 TOP 20 CIVILIZATIONMOSTSELECTED with Sum of ACTIVEDAYS 
df3 = df.groupby(by = ['CIVILIZATIONMOSTSELECTED']).ACTIVEDAYS.sum()
df3 = df3.sort_values('index', ascending=False)
figline1=plt.figure()
figbar =plt.figure()
ax1 = figbar.add_subplot(1,1,1)
ax1.set_xlabel('CIVILIZATIONMOSTSELECTED')
ax1.set_ylabel('Sum of ACTIVEDAYS')
ax1.set_title("TOP 20 CIVILIZATIONMOSTSELECTED with  Sum of ACTIVEDAYS")
df3.head(20).plot(kind='bar')
```

```python
##Question 4 :: select Top 9 England Nations CIVILIZATIONMOSTSELECTED with  Sum of ACTIVEDAYS
listofEnglandselections = df['CIVILIZATIONMOSTSELECTED'] =='CIVILIZATION_ENGLAND'
df4 = df[listofEnglandselections]
##meanactivedays = df4['ACTIVEDAYS']  > df4['ACTIVEDAYS'].mean()
df5 = df4 ##[meanactivedays]
df6 = df5.groupby('LEADERMOSTSELECTED').ACTIVEDAYS.sum()
df6 = df6.sort_values('index', ascending=False)
figline1=plt.figure()
##print(df6)
ax2 = figline1.add_subplot(1,1,1)
ax2.set_xlabel('LEADERMOSTSELECTED')
ax2.set_ylabel('Sum of ACTIVEDAYS')
ax2.set_title("Top 9 England Nations CIVILIZATIONMOSTSELECTED with  Sum of ACTIVEDAYS")
df6.head(9).plot(kind = 'bar')
```

```python
##Question 5::: select Active dates LEADER_GANDHI  was selected by the players
import matplotlib.pyplot as plt
listofGandhi   = df['LEADERMOSTSELECTED'] =='LEADER_GANDHI'
dfg = df[listofGandhi]
figline1=plt.figure()
plt.ylim(0, 1000)
axg = figline1.add_subplot(1,1,1)
axg.set_xlabel('LEADERMOSTSELECTED')
axg.set_ylabel('LASTACTIVEDAY  of Leader Gandhi')
axg.set_title("Active dates LEADER_GANDHI  was selected by the players")
##dfg.plot(kind = 'bar')
df7 = dfg.groupby('LASTACTIVEDAY').ACTIVEDAYS.sum()
df7.plot(kind = "line")

```

```python
##Question 6 ::: select top 10 active days wise selected leaders
dfindate = df.groupby(['LEADERMOSTSELECTED']).ACTIVEDAYS.sum()
temp = dfindate.sort_values('index', ascending=False)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.pie(x= temp.head(10),labels = temp.head(10).index)
plt.title("Top 10 Active days wise selected leaders")
plt.show()
```

```python
##Question 7 : Selec the active civilizations users selected  on the day of "09/01/2019"
import seaborn as sns
allin2019 = df['LASTACTIVEDAY'] == '09/01/2019'
df2 = df[allin2019]
sns.barplot(x="ACTIVEDAYS", y='CIVILIZATIONMOSTSELECTED', data = df2.head(5))
```
